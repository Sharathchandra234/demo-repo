```python
# CODE CELL 4: Core Functions and Metrics (R2 GUARDRAIL FIX)

# --- 4.1 Hypothesis Function ---
def hypothesis(X, theta):
    return X @ theta

# --- 4.2 Cost Function (Mean Squared Error) ---
def compute_cost(X, y, theta):
    m = len(y)
    predictions = hypothesis(X, theta)
    errors = predictions - y
    cost = (1 / (2 * m)) * np.sum(errors**2) 
    return cost

# --- 4.3 Evaluation Metrics (FIXED R2) ---
def evaluate_metrics(y_actual, y_predicted):
    """Calculates MSE, RMSE, MAE, and R-squared with zero-variance check."""
    
    # Ensure y_actual and y_predicted are NumPy arrays for calculations
    y_actual = np.asarray(y_actual)
    y_predicted = np.asarray(y_predicted)
    
    # Basic Metrics
    mse = np.mean((y_predicted - y_actual)**2)
    rmse = np.sqrt(mse)
    mae = np.mean(np.abs(y_predicted - y_actual))
    
    # R-squared (R2) Calculation
    sst = np.sum((y_actual - np.mean(y_actual))**2) # Total Sum of Squares
    
    # CRITICAL FIX: Check if SST is zero or near zero (no variance in actual data)
    if sst < 1e-8: # Use a small tolerance for floating point comparison
        ssr = np.sum((y_actual - y_predicted)**2)
        if ssr < 1e-8:
             # Perfect prediction on constant data
             r2 = 1.0 
        else:
             # Undefined, model has error on constant data
             r2 = np.nan 
    else:
        # Standard R2 calculation
        ssr = np.sum((y_actual - y_predicted)**2) # Residual Sum of Squares
        r2 = 1 - (ssr / sst)
    
    return {'MSE': mse, 'RMSE': rmse, 'MAE': mae, 'R2': r2}

# --- 4.4 Denormalization Function ---
def denormalize(y_norm, mean, std):
    return y_norm * std + mean
```

```python
# CODE CELL 5: Gradient Descent Algorithms

# --- 5.1 Batch Gradient Descent (BGD) ---
def batch_gradient_descent(X, y, theta, learning_rate, n_iterations, X_val=None, y_val=None):
    """
    Trains the model using Batch Gradient Descent.
    Updates parameters using ALL samples in the training set.
    """
    m = len(y)
    J_history = []
    J_val_history = []

    for i in range(n_iterations):
        # Calculate predictions
        predictions = hypothesis(X, theta)
        # Calculate error
        errors = predictions - y
        
        # Calculate the gradient: (1/m) * X^T * errors
        gradient = (1/m) * X.T @ errors
        
        # Update theta: theta = theta - alpha * gradient
        theta = theta - learning_rate * gradient
        
        # Record training and validation cost
        J_history.append(compute_cost(X, y, theta))
        if X_val is not None:
            J_val_history.append(compute_cost(X_val, y_val, theta))
            
    return theta, J_history, J_val_history

# --- 5.2 Stochastic Gradient Descent (SGD) ---
def stochastic_gradient_descent(X, y, theta, learning_rate, n_epochs, X_val=None, y_val=None):
    """
    Trains the model using Stochastic Gradient Descent (SGD).
    Updates parameters using ONE sample at a time.
    """
    m = len(y)
    J_history = []
    J_val_history = []
    
    for epoch in range(n_epochs):
        # Loop through each sample in the dataset
        for i in range(m):
            # Select a single random sample (or just iterate through them)
            # Choosing a random index for true stochastic behavior
            rand_index = np.random.randint(m)
            xi = X[rand_index:rand_index+1] # (1, n+1)
            yi = y[rand_index:rand_index+1] # (1, 1)

            # Calculate error for this single sample
            prediction = hypothesis(xi, theta)
            error = prediction - yi
            
            # Calculate the gradient for this single sample: xi.T * error
            # Note: The (1/m) factor is typically omitted/absorbed into the learning rate for SGD
            gradient = xi.T @ error 
            
            # Update theta
            theta = theta - learning_rate * gradient
            
        # Record cost at the end of each epoch
        J_history.append(compute_cost(X, y, theta))
        if X_val is not None:
            J_val_history.append(compute_cost(X_val, y_val, theta))
            
    return theta, J_history, J_val_history
```

```python
# CODE CELL 8: K-Means Clustering Implementation

### --- 1. K-Means Helper Functions ---

def euclidean_distance(point1, point2):
    """Calculates the Euclidean distance between two points."""
    return np.sqrt(np.sum((point1 - point2)**2))

def initialize_centroids(X, K):
    """Randomly selects K data points as initial centroids."""
    m, n = X.shape
    # Select K random indices from the data
    random_indices = np.random.choice(m, K, replace=False)
    # Centroids are set to the data points at those indices
    centroids = X[random_indices]
    return centroids

def assign_clusters(X, centroids):
    """Assigns each data point to the nearest centroid."""
    m = X.shape[0]
    K = centroids.shape[0]
    
    # Array to store the cluster index for each data point
    clusters = np.zeros(m, dtype=int)
    
    for i in range(m):
        min_distance = float('inf')
        min_index = -1
        for k in range(K):
            dist = euclidean_distance(X[i], centroids[k])
            if dist < min_distance:
                min_distance = dist
                min_index = k
        clusters[i] = min_index
        
    return clusters

def update_centroids(X, clusters, K):
    """Calculates the new mean position for each cluster."""
    m, n = X.shape
    new_centroids = np.zeros((K, n))
    
    for k in range(K):
        # Find all data points assigned to cluster k
        points_in_cluster = X[clusters == k]
        
        if len(points_in_cluster) > 0:
            # Calculate the mean of those points
            new_centroids[k] = np.mean(points_in_cluster, axis=0)
        else:
            # If a cluster is empty, reinitialize its centroid (a simple strategy)
            new_centroids[k] = X[np.random.randint(m)]
            
    return new_centroids

def calculate_inertia(X, centroids, clusters):
    """Calculates the WCSS (Within-Cluster Sum of Squares) or Inertia."""
    inertia = 0
    for k in range(centroids.shape[0]):
        points_in_cluster = X[clusters == k]
        for point in points_in_cluster:
            inertia += euclidean_distance(point, centroids[k])**2
    return inertia

### --- 2. K-Means Main Algorithm ---

def kmeans(X, K, max_iter=100, tolerance=1e-4):
    """Main K-Means algorithm function."""
    
    # 1. Initialize centroids
    centroids = initialize_centroids(X, K)
    
    # Storage for tracking inertia (loss)
    inertia_history = []
    
    for iteration in range(max_iter):
        # 2. Assign clusters
        clusters = assign_clusters(X, centroids)
        
        # 3. Update centroids
        new_centroids = update_centroids(X, clusters, K)
        
        # 4. Calculate inertia
        current_inertia = calculate_inertia(X, new_centroids, clusters)
        inertia_history.append(current_inertia)
        
        # 5. Check for convergence (centroids stop moving)
        centroid_movement = euclidean_distance(centroids, new_centroids)
        
        centroids = new_centroids
        
        if centroid_movement < tolerance:
            # print(f"K-Means converged in {iteration + 1} iterations.")
            break
            
    return clusters, centroids, inertia_history
```

```python
# CODE CELL 9: Elbow Method and Optimal K Selection

# Use the normalized feature data (excluding the bias term X_final[:, 0])
# X_norm was defined in Code Cell 2
if 'X_norm' in locals():
    X_clustering = X_norm.copy()
    
    # Define a range of K values to test
    K_range = range(1, 11)
    inertia_values = []
    
    print("Running K-Means for Elbow Method...")
    
    for K in K_range:
        # Run K-Means multiple times to avoid bad local minima
        best_inertia = float('inf')
        
        for run in range(5): 
            clusters, centroids, _ = kmeans(X_clustering, K)
            current_inertia = calculate_inertia(X_clustering, centroids, clusters)
            if current_inertia < best_inertia:
                best_inertia = current_inertia
                
        inertia_values.append(best_inertia)

    # --- Visualization 1: Elbow Plot (Required for Task 2) ---
    plt.figure(figsize=(10, 6))
    plt.plot(K_range, inertia_values, marker='o', linestyle='-', color='indigo')
    plt.title('Elbow Method for Optimal K (Inertia vs. Number of Clusters)')
    plt.xlabel('Number of Clusters (K)')
    plt.ylabel('Inertia (WCSS)')
    plt.xticks(K_range)
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()
    # 
    
    # Based on the plot, typically 3 or 4 is chosen for demographic data
    OPTIMAL_K = 3 
    print(f"\nOptimal K selected: {OPTIMAL_K}")
    
    # Final training with the optimal K
    final_clusters, final_centroids, _ = kmeans(X_clustering, OPTIMAL_K)
    
    # Attach cluster assignments to the original clean DataFrame for interpretation
    df_cleaned['Cluster'] = final_clusters
```
